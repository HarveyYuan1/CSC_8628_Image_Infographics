## numpy

### The Basics

```python
import numpy as np

a = np.array([2,3,4])
print(a)
print(a.dtype)
print(a.shape)
print(a.size)
```

### Array Creation

```python
a = np.zeros((3, 4))
print(a)
print(a.shape)

a = np.ones((3, 4))
print(a)
```

```python
a = np.ones((2,3,4), dtype=np.int16)
print(a)
print(a.dtype)
```

### Basic Operations

```python
a = np.array([20,30,40,50])
b = np.arange(4)
print(a)
print(b)

c = a-b
print(c)

d = a<35
print(d)
```

```python
a = np.array([[1,1], [0,1]])
b = np.array([[2,0], [3,4]])
print(a)
print(b)

c = a*b # elementwise product
print(c)
```

### Universal Functions

```python
a = np.arange(3)
print(a)

b = np.exp(a)
print(b)

c = np.sqrt(a)
print(c)
```

### Indexing, Slicing and Iterating

```python
a = np.arange(10)**3
print(a)
print(a[2])
print(a[2:5])

a[:6:2] = 1000
print(a)

print(a[ : :-1])

for i in a:
  print(i**(1/3.))
```

```python
a = np.random.rand(5,4)
print(a)
print(a[2,2])
print(a[0:2,2])
print(a[:,2])
print(a[1:3,:])
print(a[-1])
```

### Indexing with Boolean Arrays

```python
a = np.arange(12).reshape(3,4)
b = a > 4
print(b)

c = a[b]
print(c)
```

### Multiple figures and axes

```python
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

### Working with text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

# Task: Read and display image

Use the function cv2.imread() to read an image. The image should be in the working directory, or a full path of
image should be given.

Second argument is a flag which specifies the way image should be read.

cv2.IMREAD_COLOR: Loads a color image. Any transparency of image will be neglected. It is the default flag.

cv2.IMREAD_GRAYSCALE: Loads image in grayscale mode

cv2.IMREAD_UNCHANGED: Loads image as such including alpha channel

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
# Load an color image
img_color = cv2.imread('P1-images/NU.jpg', cv2.IMREAD_COLOR)
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

plt.imshow(img_color)
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```

```python
# Load an color image in grayscale
img_gray = cv2.imread('P1-images/NU.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```
